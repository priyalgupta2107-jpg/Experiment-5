```python
# Experiment 5
# Study of Dictionaries in python

# Creation of Dictionary
thisdict = {
    "Brand": "Ford",
  "Model": "Mustang",
  "Year": 1964
}
print(thisdict)

```

```python
# retriving values from Dictionary
thisdict = {
    "Brand": "Ford",
  "Model": "Mustang",
  "Year": 1964
}
print(thisdict["Brand"])
```

```python
# Dictionary takes latest value given for key. it will not allow duplicates
thisdict = {
    "Brand": "Ford",
  "Model": "Mustang",
  "Year": 1964,
  "Year": 2020
}
print(thisdict)

# Length of key value pairs
len(thisdict)
```

```python
# Changing/Adding key:value in dictionary
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}
car["color"] = "red"
print(car)
x = car.values()
print(x)
print(type(car))
```

```python
# Q3).A dictionary stores usernames and passwords.Write a python program to
# validate user login.

users = {"admin":'1234', "student": "abcd"}
user_name = input("Username:")
pw= input("Password:")

if users.get(user_name) == pw:
  print("Login Successful")
else:
  print("Invalid Username or password")
```

```python
# Q4).A dictionary stores student names and marks.
# Write a python program to find the student who scores the highest marks and
# print the name with the marks

marks ={"A":85, "B":92,"C":78}
topper = max(marks, key=marks.get)
print("Topper:",topper)
print("Marks:",marks[topper])
```
